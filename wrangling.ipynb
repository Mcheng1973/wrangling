```python
url = 'https://raw.githubusercontent.com/Mcheng1973/wrangling/refs/heads/main/assignment/data/sharks.csv'
df = pd.read_csv(url)
df['Type'].value_counts()
```

```python
type = df['Type']

type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft')
type.value_counts()

type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan)
type.value_counts()

df['Type'] = type
del type

df['Type'].value_counts()
```

```python
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # All unclean values
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y')
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')
```
